# Analysis on the first 1 million digits of Pi
This program answers the following questions regarding the first million digits of Pi:

1. What is the most and least frequent digit?
2. What are the most frequent same-digit consecutive numbers?
3. Statistical summary (average, standard deviation, etc.)

You can read more about Pi [here](https://en.wikipedia.org/wiki/Pi)

```python
from matplotlib import pyplot as plt
```

## Store `pi.txt` in a variable

```python
# 1 Million digits of Pi
with open('pi.txt', 'r') as file:
    pi = file.read()
    print(len(pi))
```

## Store in a dictionary the frequency of each digit (**excluding the dot**)

```python
# Count of each digit
digit_count = {}
pi_no_dot = pi.replace('.', '')
for char in pi_no_dot:
    if char not in digit_count:
        digit_count[char] = 1
        continue
    digit_count[char] += 1
print(digit_count)
```

## Set a digits and counts array, then sort them

```python
digits = digit_count.keys()
counts = digit_count.values()
print(digits, counts, end='\n')
digits, counts = (list(t) for t in zip(*sorted(zip(digits, counts))))
print(digits, counts, end='\n')
```

## Plot the digits frequency using `matplotlib`

```python
plt.plot(digits, counts, marker='o', color='red', markeredgecolor='black')
plt.title('Digits count in the first 1 million digits of Pi', fontfamily='Monospace')
plt.xlabel('Digits of Pi', fontfamily='Monospace')
plt.ylabel('Frequency', fontfamily='Monospace')

# plt.yticks(list(plt.yticks()[0]) + counts)
plt.yticks(counts, fontsize=8, fontfamily='Verdana')
plt.tight_layout()

plt.grid()

plt.show()
```

Here we can see that the most frequent digit (`f = 100,358`) is 5, and the least frequent being 6 (`f = 99,548`).
